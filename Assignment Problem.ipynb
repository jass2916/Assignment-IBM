```python
# Employee Salary Calculator
def calculate_hra(basic_salary):
    return basic_salary * 0.20

def calculate_da(basic_salary):
    return basic_salary * 0.10

def calculate_net_salary(basic_salary):
    hra = calculate_hra(basic_salary)
    da = calculate_da(basic_salary)
    return basic_salary + hra + da

# Input
name = input("Enter employee name: ")
basic = float(input("Enter basic salary: "))

# Calculation
hra = calculate_hra(basic)
da = calculate_da(basic)
net = calculate_net_salary(basic)

# Display
print(f"\n--- Salary Breakdown for {name} ---")
print(f"Basic Salary: {basic:.2f}")
print(f"HRA (20%):    {hra:.2f}")
print(f"DA (10%):     {da:.2f}")
print(f"Net Salary:   {net:.2f}")
```

```python
# Student Grade Management System
def calculate_total(marks):
    return sum(marks)

def calculate_percentage(total, num_subjects):
    return (total / (num_subjects * 100)) * 100

def assign_grade(percentage):
    if percentage >= 90: return 'A'
    elif percentage >= 75: return 'B'
    elif percentage >= 60: return 'C'
    elif percentage >= 40: return 'D'
    else: return 'F'

# Input
marks = []
for i in range(1, 6):
    marks.append(float(input(f"Enter marks for subject {i}: ")))

# Processing
total = calculate_total(marks)
percent = calculate_percentage(total, 5)
grade = assign_grade(percent)

# Output
print(f"\nTotal Marks: {total}")
print(f"Percentage:  {percent:.2f}%")
print(f"Final Grade: {grade}")
```

```python
# Student Performance Management System
# Grade criteria as a tuple
GRADES = ('A', 'B', 'C', 'D', 'F')

def get_grade(percentage):
    if percentage >= 90: return GRADES[0]
    elif percentage >= 75: return GRADES[1]
    elif percentage >= 60: return GRADES[2]
    elif percentage >= 40: return GRADES[3]
    else: return GRADES[4]

def process_students():
    students = []
    n = int(input("Enter number of students: "))

    for _ in range(n):
        record = {}
        record['roll'] = input("\nEnter Roll No: ")
        record['name'] = input("Enter Name: ")
        record['marks'] = [float(input(f"Marks for subject {i+1}: ")) for i in range(5)]

        # Calculations
        record['total'] = sum(record['marks'])
        record['percent'] = (record['total'] / 500) * 100
        record['grade'] = get_grade(record['percent'])

        students.append(record)

    # Display Records
    print("\n" + "="*30)
    print("STUDENT RECORDS")
    print("="*30)
    for s in students:
        print(f"Roll: {s['roll']} | Name: {s['name']} | Total: {s['total']} | %: {s['percent']:.2f} | Grade: {s['grade']}")

    # Find Topper
    topper = max(students, key=lambda x: x['percent'])
    print(f"\nTopper: {topper['name']} with {topper['percent']:.2f}%")

process_students()
```
