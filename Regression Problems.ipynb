```python
# Students Performance Analysis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv('StudentsPerformance.csv')

# Check for missing values
missing_values = df.isnull().sum()

# Calculations using NumPy for 'math score'
marks = df['math score'].values
mean_marks = np.mean(marks)
median_marks = np.median(marks)
std_marks = np.std(marks)

print(f"Mean: {mean_marks:.2f}")
print(f"Median: {median_marks}")
print(f"Standard Deviation: {std_marks:.2f}")
```

```python
# Histogram of Marks
plt.figure(figsize=(8, 4))
plt.hist(df['math score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Math Scores')
plt.xlabel('Marks')
plt.ylabel('Frequency')
plt.savefig('histogram_marks.png')

# Scatter Plot (Using Reading Score as proxy for Study Hours)
plt.figure(figsize=(8, 4))
plt.scatter(df['reading score'], df['math score'], alpha=0.5)
plt.title('Reading Score vs Math Score')
plt.xlabel('Reading Score')
plt.ylabel('Math Score')
plt.savefig('scatter_plot.png')
```

```python
# Prepare data
X = df[['reading score']].values
y = df['math score'].values

# Build Model
model = LinearRegression()
model.fit(X, y)
predictions = model.predict(X)

# Display Regression Line
plt.figure(figsize=(8, 4))
plt.scatter(X, y, alpha=0.5, label='Data points')
plt.plot(X, predictions, color='red', linewidth=2, label='Regression Line')
plt.title('Linear Regression: Reading Score vs Math Score')
plt.xlabel('Reading Score')
plt.ylabel('Math Score')
plt.legend()
plt.savefig('regression_line.png')
```

```python
# Car Mileage Prediction
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Read dataset
# Using appropriate delimiter if necessary for UCI datasets
df_car = pd.read_csv('/content/auto-mpg.data', delim_whitespace=True, header=None)
df_car.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model year', 'origin', 'car name']

# 2. Descriptive statistics
print(df_car.describe())

# 3. Plots
plt.scatter(df_car['weight'], df_car['mpg'])
plt.title('Weight vs Mileage')
plt.show()

plt.hist(df_car['mpg'], bins=15)
plt.title('Histogram of Mileage')
plt.show()

# 4 & 5. Linear Regression (Engine Size/Displacement as proxy for Engine Size)
X_car = df_car[['displacement']].values
y_car = df_car['mpg'].values
model_car = LinearRegression().fit(X_car, y_car)
pred_car = model_car.predict(X_car)

plt.scatter(X_car, y_car)
plt.plot(X_car, pred_car, color='red')
plt.title('Regression: Engine Size vs Mileage')
plt.show()
```
