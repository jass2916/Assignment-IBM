```python
# Student Pass/Fail Analysis
passed = []
failed = []

with open('/content/drive/MyDrive/Colab Notebooks/student.txt', 'r') as file:
    for line in file:
        name, marks = line.strip().split(',')
        if int(marks) > 45:
            passed.append(name)
        else:
            failed.append(name)

print(f"Passed Students: {', '.join(passed)} (Count: {len(passed)})")
print(f"Failed Students: {', '.join(failed)} (Count: {len(failed)})")
```

```python
# Employee Salary Filter
count = 0
print("Employees with salary > ₹50,000:")

with open('/content/drive/MyDrive/Colab Notebooks/Salary.txt', 'r') as file:
    for line in file:
        name, salary = line.strip().split(',')
        if int(salary) > 50000:
            print(f"- {name}")
            count += 1

print(f"\nTotal employees earning > ₹50,000: {count}")
```

```python
# Attendance Tracker
present_count = 0
absent_count = 0
absent_students = []

with open('/content/drive/MyDrive/Colab Notebooks/Attendance.txt', 'r') as file:
    for line in file:
        name, status = line.strip().split(',')
        if status.lower() == 'present':
            present_count += 1
        else:
            absent_count += 1
            absent_students.append(name)

print(f"Absent Students: {', '.join(absent_students)}")
print(f"Total Present: {present_count}")
print(f"Total Absent: {absent_count}")
```
