```python
import pandas as pd
import numpy as np

# 1. Create a sample CSV file
data = {
    'Product': ['Laptop', 'Mouse', 'Monitor', 'Keyboard', 'Headphones'],
    'Price': [800, 20, 150, 40, 60]
}
df = pd.DataFrame(data)
df.to_csv('products.csv', index=False)

# 2. Read the data
df_read = pd.read_csv('products.csv')

# 3. Calculate average price using NumPy
avg_price = np.mean(df_read['Price'])

# 4. Function to identify products above average
def get_expensive_products(df, average):
    print(f"Average Price: ${average:.2f}\n")
    print("Products priced above average:")
    # Filter using boolean indexing
    expensive_items = df[df['Price'] > average]
    print(expensive_items)

get_expensive_products(df_read, avg_price)
```

```python
import pandas as pd
import numpy as np

# 1. Create sample temperature CSV
temp_data = {
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Temperature': [15, 25, 35, 18, 22]
}
df_temp = pd.DataFrame(temp_data)
df_temp.to_csv('temperatures.csv', index=False)

# 2. Read data
df = pd.read_csv('temperatures.csv')

# 3. Use NumPy for statistics
high = np.max(df['Temperature'])
low = np.min(df['Temperature'])
avg = np.mean(df['Temperature'])

print(f"Max: {high}°C, Min: {low}°C, Avg: {avg:.2f}°C\n")

# 4. Function to classify temperature
def classify_temp(temp):
    if temp < 20:
        return 'Cold'
    elif 20 <= temp <= 30:
        return 'Moderate'
    else:
        return 'Hot'

# Apply the function to the DataFrame
df['Category'] = df['Temperature'].apply(classify_temp)

print("Daily Temperature Classification:")
print(df)
```
